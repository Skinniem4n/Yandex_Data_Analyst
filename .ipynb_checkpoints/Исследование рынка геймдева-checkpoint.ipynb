```python
# Приведем названия столбцов к нижнему регистру с помощью str.lower()

data.columns = data.columns.str.lower()
```

```python
# И сделаем то же самое для строковых стоолбцов

data['name'] = data['name'].str.lower()
data['platform'] = data['platform'].str.lower()
data['genre'] = data['genre'].str.lower()
```

```python
data.head()
```

Посмотрим, какие платформы вообще рассматриваются в датасете.

```python
data['platform'].unique()
```

Я принял решение убрать данные с пропущенными названиями самой игры. Таких строки всего две, на исследование они не повлияют.

```python
data = data.dropna(subset=['name'])
```

Приведем оценки критиков и юзеров к одному формату, чтобы было удобнее сравнивать.

```python
data['critic_score'] = data['critic_score'] / 10
```

```python
data.info()
```

Довольно много пропусков в оценках от критиков и юзеров. Это может быть вызвано банальным отсутствием данных по поводу рейтинга тех или иных игр ввиду их возраста или чего-то еще. Также информация об оценках могла не подтянуться в базу.

В любом случае, заполним пропуски в этих столбцах единым числовым значением. Ноль не подойдет, можно спутать с нулевой оценкой. А вот -1, как мне кажется, подходит отлично.

```python
data['critic_score'] = data['critic_score'].fillna(-1.0)
```

```python
data['user_score'] = data['user_score'].fillna(-1.0)
```

```python
data.info()
```

Очень долго думал, как быть со значением `tbd` в столбце `user_score`. Потом понял очевидное: `to be determined`, по сути, означает, что данных нет. Проставим пропуски!

```python
data[data['user_score'] == 'tbd']
```

```python
# Заполним NaN'ами значения 'tbd' в столбце 'user_score' с помощью метода np.NaN

data.loc[data['user_score'] == 'tbd','user_score'] = np.NaN
```

Отлично, осталось только привести столбец к типу float, чтобы было удобнее работать с ним в дальнейшем.

```python
data['user_score'] = data['user_score'].astype(float)
```

Заполнение пропусков в дате релиза числом 2077, если честно, подсмотрел, когда изучал гиты студентов этого же курса. Ничего не смог с собой поделать, очень уж красивое решение и отличная отсылка.

Заполнить оказалось необходимо, чтобы привести столбец к целочисленному типу. Пропусков тут не так много, допускаю человеческий фактор (могли просто забыть указать).

```python
data['year_of_release'] = data['year_of_release'].fillna(2077)
```

```python
data['year_of_release'] = data['year_of_release'].astype(int)
```

```python
data.head()
```

```python
data.info()
```

Добавим столбец с суммарными продажами. Для этого сложим значения 4 столбцов со сборами в разных регионах.

```python
data['total_sales'] = data[['na_sales','eu_sales','jp_sales', 'other_sales']].sum(axis = 1)
```

```python
data.head()
```

```python
# Посмотрим, какие рейтинги бывают у игр

data['user_score'] = data['user_score'].astype(float)
```

```python
data['rating'].unique()
```

Также добавим столбец с расшифровкой рейтинга, обозначения не понятны интуитивно.

# Проведем исследовательский анализ

Для начала посмотрим, сколько игр выпускалось в разные годы, чтобы в дальнейшем определить актуальный период, с которым мы будем работать в нашем исследовании.

```python
data['year_of_release'].hist(bins=37, range=(1980, 2017), color='tomato')
plt.title('Кол-во выпущенных игр в разные годы')
plt.xlabel('Годы')
plt.ylabel('Кол-во выпущенных игр');
```

В период с 1980 по 1990 игр выпускалось не так много, расцвет игровой индустрии начался в середине 90-х, а во второй половине 2000-х индустрия, судя по всему, достигла своего пика.

На первый взгляд может показаться, что рынок геймдева находится в драматических масштабов упадке с начала 2010-х годов. Не будем спешить с выводами, для начала посмотрим, продажи игр для каких платформ представлены в датасете больше остальных.

```python
# Посмотрим на 10 платформ с наибольшими продажами

sorted_platform = data.groupby('platform').agg({'total_sales': 'sum'})
sorted_platform.sort_values(by='total_sales', ascending=False).head(10)
```

Снова видим преобладание платформ, популярных в 2000-х годах, тогда как ПК и, например PlayStation 4 находятся внизу нашего топа.

Вспоминая, что в датасете представлены данные за 2016 год, могу предположить, что то, что на графике выше выглядит как упадок, могло сложиться по двум причинам:
- С момента выхода приставок нового поколения прошло всего 3 года, и именно с этим связано небольшое количество продаж после 2010-х годов: достаточное количество игр просто не успело выйти, а для приставок старого поколения их выпускать постепенно перестали;
- В датасете не учтены мобильные платформы (смартфоны, планшеты и т.п.), упор на которые сейчас и делается в игровой индустрии.

Раз мы выбрали отдельный актуальный период, думаю лучше всего будет создать отдельный датасет на основе уже существующего, срезав год релиза.

```python
# Создаем новый датафрейм, и выведем случайные 5 строк, чтобы убедиться, что всё ок

games = data[data['year_of_release'] >= 2013].reset_index(drop=True)
games.sample(5)
```

По аналогии с тем, как делали раньше, посмотрим самые "продающие" платформы, но теперь за актуальный период.

```python
sorted_platform_actual = games.groupby('platform').agg({'total_sales': 'sum'})
sorted_platform_actual.sort_values(by='total_sales', ascending=False).head(10)
```

Дополним информацией по годам. Для этого построим график и посмотрим, какие платформы лидируют по продажам в последнее время. Это необходимо, чтобы понимать, как платформы сейчас обладают наибольшим потенциалом.

```python
# Создадим сводную таблицу и на ее основе сразу построим график

games_pivot = pd.pivot_table(games[games['year_of_release'] < 2077],\
                             index='year_of_release',\
                             columns='platform',\
                             values='total_sales',\
                             aggfunc='sum').plot(figsize=(18, 10), grid=True)

plt.title('Продажи на разных платформах за период 2013-2016 гг', fontsize=15)
plt.xlabel('Год релиза', fontsize=15)
plt.ylabel('Продажи, млн', fontsize=15);
```

С большим отрывом прямо сейчас лидирует PlayStation 4, далее идет ее прямой конкурент - Xbox One, замыкает тройку 3DS. Это не удивительно, ведь все три платформы представляют новейшее поколение консолей (на 2016 год). Однако, на этом графике мы особенно четко видим тенденцию к снижению продаж на всех платформах, включая упомянутые выше.

Не стоит удивляться четвертому месту ПК - игры на компьютер стоят дешевле игр на консоли. Не будем также забывать про пиратство и огромное количество f2p проектов.

```python
games['total_sales'].describe()
```

Построим "ящик с усами", отражающий глобальные продажи в разбивке по платформам.

```python
games.query('2016 >= year_of_release > 2013').boxplot(column = ['total_sales'], by = 'platform', figsize=(15, 6))
plt.title('Глобальные продажи по платформам с выбросами', fontsize=15);
```

```python
games.query('2016 >= year_of_release > 2013').boxplot(column = ['total_sales'], by = 'platform', figsize=(15, 6))
plt.title('Глобальные продажи по платформам без выбросов', fontsize=15)
plt.ylim(0, 1.6);
```

На первом графике видно, что выбросов больше всего у консолей новейшего поколения: PS4, Xbox One и 3DS. Выглядит логично, учитывая, что на новый консоли должно выходить больше всего ААА-блокбастеров, которые и являются выбросами.

# Составим портрет пользователя каждого региона

Определим для каждого пользователя трёх рассматриваемых нами рынков:
- Топ-5 популярных платформ;
- Топ-5 популярных жанров;
- Влияние рейтинга на продажи.

Построим для каждого пункта по три графика (по одному на рынок), сделаем выводы. В конце напишем общий вывод.

```python
def platform_categorizer_na_eu(row):
    if row in ['ps4', 'ps3', 'xone', 'x360', '3ds']:
        return row
    else:
        return 'other'

def platform_categorizer_jp(row):
    if row in ['ps4', 'ps3', '3ds', 'wiiu', 'psv']:
        return row
    else:
        return 'other'
    
games['platform_na_top'] = games['platform'].apply(platform_categorizer_na_eu)
games['platform_eu_top'] = games['platform'].apply(platform_categorizer_na_eu)
games['platform_jp_top'] = games['platform'].apply(platform_categorizer_jp)

na_platforms = games.groupby('platform_na_top').agg({'na_sales': 'sum'})
eu_platforms = games.groupby('platform_eu_top').agg({'eu_sales': 'sum'})
jp_platforms = games.groupby('platform_jp_top').agg({'jp_sales': 'sum'})
```

В Японии отдают предпочтение своим платформам. Доминирует 3DS, причем с огромным отрывом, далее идут продукты Sony, Wii U, и только потом все остальные. Выходить на японский рынок точно стоит только с японскими продуктами :)

```python
def genre_categorizer_na(row):
    if row in ['shooter', 'action', 'sports', 'misc']:
        return row
    elif row == 'role-playing':
        return 'roleplay'
    else:
        return 'other'
    
def genre_categorizer_eu(row):
    if row in ['shooter', 'action', 'sports', 'racing']:
        return row
    elif row == 'role-playing':
        return 'roleplay'
    else:
        return 'other'
    
def genre_categorizer_jp(row):
    if row in ['action', 'fighting', 'misc', 'shooter']:
        return row
    elif row == 'role-playing':
        return 'roleplay'
    else:
        return 'other'
    
games['genre_na_top'] = games['genre'].apply(genre_categorizer_na)
games['genre_eu_top'] = games['genre'].apply(genre_categorizer_eu)
games['genre_jp_top'] = games['genre'].apply(genre_categorizer_jp)

na_genres = games.groupby('genre_na_top').agg({'na_sales': 'sum'})
eu_genres = games.groupby('genre_eu_top').agg({'eu_sales': 'sum'})
jp_genres = games.groupby('genre_jp_top').agg({'jp_sales': 'sum'})
```

# Проверим гипотезы

- Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
- Средние пользовательские рейтинги жанров Action и Sports разные.

Для проверки гипотез будем использовать изначальный датасет `data`, чтобы рассмотреть данные за всю историю существования игр, т.к. в задании не указано обратного.

Проверим гипотезы с помощью метода `ttest_ind()`, потому что выборки независимы.

- Гипотеза Н0: средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
- Гипотеза Н1: средние пользовательские рейтинги платформ Xbox One и PC различаются.

```python
alpha = 0.05

xbox_one = data[(data['platform'] == 'xone') & (data['user_score'] > 0)]['user_score']
pc = data[(data['platform'] == 'pc') & (data['user_score'] > 0) & (data['year_of_release'] >= 2013)]['user_score']

results = st.ttest_ind(xbox_one, pc)
print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

- Гипотеза Н0: cредние пользовательские рейтинги жанров Action и Sports одинаковые;
- Гипотеза Н1: cредние пользовательские рейтинги жанров Action и Sports различаются.

```python
alpha = 0.05

action = data[(data['genre'] == 'action') & (data['user_score'] > 0)]['user_score']
sports = data[(data['genre'] == 'sports') & (data['user_score'] > 0)]['user_score']

results = st.ttest_ind(action, sports)
print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

У нас нет оснований утверждать, что выдвинутые гипотезы доказаны, основываясь на проверке выше. Однако же, нам не удалось отвергнуть следующие гипотезы:

- Средние пользовательские рейтинги платформ Xbox One и PC (если считать актуальный период с 2013 года) не различаются;
- Средние пользовательские рейтинги жанров Action и Sports одинаковые.
